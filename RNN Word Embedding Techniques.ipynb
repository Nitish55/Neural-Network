```python
# importng libraries
import tensorflow as tf
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot
import numpy as np
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
sent
```

**Initialising the vocubalary**

```python
### Vocabulary size
voc_size=10000
```

## One Hot Representation

```python
onehot_repr=[one_hot(wrds,voc_size)for wrds in sent]
print(onehot_repr)
```

* 'the glass of milk'-- [9726, 2946, 3433, 1249]
* 'the glass of juice' -- [9726, 2946, 3433, 6200]

**Here we are getting the index of the word in the vocublary**

## Word Embedding Represntation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

**Pad-Sequence**
* it is used because texts in senteb=nxes are not of equal length that's why we use padding to make senteces length equal

```python
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding="pre",maxlen=sent_length)
embedded_docs
```

**Initializing Dimensions**

```python
dim=10
```

### Adding Embedding layer
* This layer helps us to convert based on dimension featureize representation

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

**Now in oredr to see how our words are converted**

```python
print(model.predict(embedded_docs))
```
